```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#sqrt_Er, sig12 sind Fitparameter
n1=1
n2=2
def fit_func(x, sqrt_Er, sig12):
    return sqrt_Er*(x-sig12)*np.sqrt(1/n1**2-1/n2**2)

Z=np.array([26,  42, 28, 47, 40, 30, 29, 22])
#K_alpha (Fe, Mo, Ni, Ag, Zr, Zn, Cu, Ti) in keV:
K_alpha=np.array([6.41,17.48,7.49,21.94,15.81,8.67,8.07,4.44])
Delta_K_alpha=np.array([0.16,0.17,0.16,0.22,0.18,0.16,0.17,0.15])
sqrt_K_alpha=np.sqrt(K_alpha)
Delta_sqrt_K_alpha= Delta_K_alpha / (2*sqrt_K_alpha) #unsere Formel für den Fehler
K_beta=np.array([7.04,19.57,8.29,24.6,17.71,9.61,8.93,4.83])
Delta_K_beta=np.array([0.21,0.16,0.18,0.12,0.16,0.17,0.18,0.15])
sqrt_K_beta=np.sqrt(K_beta)
Delta_sqrt_K_beta= Delta_K_beta / (2*sqrt_K_beta) #unsere Formel für den Fehler
```

```python
plt.figure(1, figsize=(8,8))
plt.errorbar(Z, sqrt_K_alpha, Delta_sqrt_K_alpha, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\alpha}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\alpha}$' + ' als Funktion von Z')
#plt.savefig("figures/K_alpha_vs_Z.pdf", format="pdf")

popta, pcova=curve_fit(fit_func, Z, sqrt_K_alpha,
sigma=Delta_sqrt_K_alpha)

plt.figure(2,figsize=(8,8))
plt.errorbar(Z, sqrt_K_alpha, Delta_sqrt_K_alpha, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\alpha}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\alpha}$' + ' als Funktion von Z')
plt.plot(Z, fit_func(Z,*popta))
#plt.savefig("figures/K_alpha_vs_Z_fit.pdf", format="pdf")

print("sqrt_Er=",popta[0], ",Standardfehler=",np.sqrt(pcova[0][0]))
print("sig12=",popta[1], ",Standardfehler=",np.sqrt(pcova[1][1]))
print("Er=", popta[0]**2, ",Standardfehler=",2*popta[0]*np.sqrt(pcova[0][0]))
```

```python
plt.figure(3, figsize=(8,8))
plt.errorbar(Z, sqrt_K_beta, Delta_sqrt_K_beta, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\beta}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\beta}$' + ' als Funktion von Z')
#plt.savefig("figures/K_beta_vs_Z.pdf", format="pdf")

poptb, pcovb=curve_fit(fit_func, Z, sqrt_K_beta,
sigma=Delta_sqrt_K_beta)

plt.figure(4,figsize=(8,8))
plt.errorbar(Z, sqrt_K_beta, Delta_sqrt_K_beta, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\beta}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\beta}$' + ' als Funktion von Z')
plt.plot(Z, fit_func(Z,*poptb))
#plt.savefig("figures/K_beta_vs_Z_fit.pdf", format="pdf")

print("sqrt_Er=",poptb[0], ",Standardfehler=",np.sqrt(pcovb[0][0]))
print("sig12=",poptb[1], ",Standardfehler=",np.sqrt(pcovb[1][1]))
print("Er=", poptb[0]**2, ",Standardfehler=",2*poptb[0]*np.sqrt(pcovb[0][0]))
```
